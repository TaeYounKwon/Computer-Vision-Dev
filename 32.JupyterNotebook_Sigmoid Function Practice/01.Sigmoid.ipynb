### 시그모이드 함수 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# [-10 ~ 10] 구간에서 100개의 t값을 sigmoid function에 대입합니다.
t = np.linspace(-10, 10, 1000)
```

```python
# sigmoid 공식
sig = 1 / (1 + np.exp(-t)) # np.exp --> 함수 밑(base)이 자연상수 e인 지수함수로 변환해주는 기능
```

```python
# t와 sigmoid의 결과값 그래프
plt.figure(figsize=(9,3))
plt.plot([-10,10], [0,0], "k-")
plt.plot([-10,10], [0.5,0.5], "k:") # y축 기준으로 0.5 실선 하나 생성
plt.plot([-10,10], [1,1], "k:") # y축 기준으로 1 실선 하나 생성
plt.plot([0,0], [-1.1,1.1], "k-") # center 기준으로 0.0  선 하나 생성
plt.plot(t, sig, "r-", linewidth=2, label=r"$\sigma(t) = \frac{1}{1+e^{-t}}$")
plt.xlabel("t")
plt.legend(loc="upper left", fontsize=20)
plt.axis([-10,10,-0.1, 1.1]) # 그래프 간격 조정
plt.show()
```

### Weight(w)값이 변함에 따라 경사도가 변화하는 그래프 시각화

```python
def sigmoid(t) :
    return 1/(1+np.exp(-t))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5 * x)
y2 = sigmoid(x)
y3 = sigmoid(2 * x)

plt.figure(figsize=(9,3))
plt.plot(x, y1, 'r', linestyle='--') # w = 0.5
plt.plot(x, y2, 'g') # w = 1
plt.plot(x, y3, 'b', linestyle='--') # w = 2
plt.plot([0,0], [1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

### Bias(b) 값이 변함에 따라 좌우 이동으로 변화하는 그래프 시각화

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5 + x)
y2 = sigmoid(1 + x)
y3 = sigmoid(1.5 + x)
```

```python
plt.figure(figsize=(9,3))
plt.plot(x, y1, 'r', linestyle='--') # b = 0.5
plt.plot(x, y2, 'g') # b = 1
plt.plot(x, y3, 'b', linestyle='--') # b = 1.5
plt.plot([0,0], [1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```
