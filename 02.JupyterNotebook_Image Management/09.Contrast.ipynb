# 히스토그램 평활화(histogram Equalization)

OpenCV에서는 equalizHist()메소드를 통해 적용 가능

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "./cat.jpg"
```

```python
# 흑백 이미지 대비 높이기
image_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
image_enhanced = cv2.equalizeHist(image_gray)
```

```python
# plot
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_gray, cmap='gray')
ax[0].set_title("Original Image")
ax[1].imshow(image_enhanced, cmap='gray')
cv2.imwrite('09.cat_black&white_contrast.png',image_enhanced)
ax[1].set_title("Enhanced Image")
plt.show()
# 흑백 이미지 대비 높이기
```

```python
# 컬러 이미지 대비 높이기
# 방법 : RGB -> YUV 컬러 포맷으로 변환 -> equlizeHist() -> RGB
# BGR
image_bgr = cv2.imread(image_path)
```

```python
# RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
# YUV
# Y는 루마 또는 밝기, U와V는 컬러를 나타냄
image_yuv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2YUV)
```

```python
# 히스토그램 평활화 적용
image_yuv[:, :, 0] = cv2.equalizeHist(image_yuv[:, :, 0])
```

```python
# RGB 변경
image_rgb_temp = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2RGB)
```

```python
# plot
fig, ax = plt.subplots(1, 2, figsize=(12, 8))
ax[0].imshow(image_rgb)
ax[0].set_title("Original Image")
ax[1].imshow(image_rgb_temp)
ax[1].set_title("Enhanced Color image")
cv2.imwrite('09.cat_color_contrast.png',image_rgb_temp)
plt.show()
```
