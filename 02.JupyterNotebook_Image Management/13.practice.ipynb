# 컬러 히스토그램을 특성으로 인코딩

이미지의 강도 분포에 대한 전반적인 아이디어를 제공하는 그래프 또는 플롯으로 히스토그램을 고려할 수
있습니다.

X축에 픽셀 값(0~255 범위, 항상 그런 것은 아님)이 있고 Y축에 이미지의 해당 픽셀 수가 있는 플롯입니다.
이미지를 이해하는 또 다른 방법일 뿐입니다.

이미지의 히스토그램을 보면 해당 이미지의 대비, 밝기, 강도 분포 등에 대한 직관을 얻을 수 있습니다. 오
늘날 거의 모든 이미지 처리 도구는 히스토그램 기능을 제공합니다.

이미지의 각 픽셀은 여러 컬러 채널(빨간, 초록, 파랑)의 조합으로 표현되며, 채널의 평균값을 계산하여
이미지의 평균 컬러를 나타내는 세 개의 컬럼 특성을 만듭니다

```python
image_bgr = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
eatures = []  # 특성 값을 담을 리스트
colors = ("r", "g", "b")  # 각 컬러 채널에 대해 히스토그램을 계산
```

```python
for i, channel in enumerate(colors):
    # cv2.calcHist([이미지], [채널 인덱스] , [마스크 없으므로 None],
    # [히스토그램 크기], [범위 : 0 ~ 256])
    histogram = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
    plt.plot(histogram, color=channel)
    plt.xlim([0, 256])
plt.savefig("13.cat_Histogram.png")
plt.show()
```
